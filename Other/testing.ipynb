```python
day1_laps = day1.laps
day1_laps = day1_laps.dropna(subset="LapTime")

day2_laps = day2.laps
day2_laps = day2_laps.dropna(subset="LapTime")

day3_laps = day3.laps
day3_laps = day3_laps.dropna(subset="LapTime")

days = [day1_laps, day2_laps, day3_laps]
```

```python
fastest_lap_day1 = day1.laps.pick_fastest()
fastest_lap_day2 = day2_laps.pick_fastest()
```

```python
drivers_day1 = pd.unique(day1.laps['Driver'])

list_fastest_laps_day1 = list()
for drv in drivers_day1:
    drvs_fastest_lap = day1.laps.pick_driver(drv).pick_fastest()
    list_fastest_laps_day1.append(drvs_fastest_lap)
fastest_laps_day1 = Laps(list_fastest_laps_day1).sort_values(by="LapTime").reset_index(drop=True)




drivers_day2 = pd.unique(day2.laps["Driver"])

list_fastest_laps_day2 = list()
for drv in drivers_day2:
    drvs_fastest_lap = day2_laps.pick_driver(drv).pick_fastest()
    list_fastest_laps_day2.append(drvs_fastest_lap)
fastest_laps_day2 = Laps(list_fastest_laps_day2).sort_values(by="LapTime").reset_index(drop=True)
```

```python
team_colours_day1 = list()
for index, lap in fastest_laps_day1.iterlaps():
    for key, value in team_colours.items():
        if key == lap["Team"]:
            colour = value
    team_colours_day1.append(colour.lower())


team_colours_day2 = list()
for index, lap in fastest_laps_day2.iterlaps():
    for key, value in team_colours.items():
        if key == lap["Team"]:
            colour = value
    team_colours_day2.append(colour.lower())
```

```python
team_long_runs = dict()
team_long_runs_filtered = dict()

for team in all_laps_by_team:
    team_data = all_laps_by_team[team]

    team_data["BoxLap"] = 0

    team_data.loc[(team_data["PitOutTime"].notnull()) | (team_data["PitInTime"].notnull()), "BoxLap"] = 1
    
    team_data["StintLength"] = team_data.groupby(["Day", "Stint"])["Stint"].transform("size")
    team_data["StintLength"] = team_data["StintLength"] - 2

    long_runs = team_data[(team_data["StintLength"] > 5) & (team_data["BoxLap"] == 0)].copy()

    team_long_runs[team] = long_runs

    # dropping stints with a high variance 
    stint_std = long_runs.groupby("Stint")["LapTime"].std().reset_index()
    long_runs_with_std = pd.merge(long_runs, stint_std, on="Stint", suffixes=("", "_std"))

    threshold = pd.Timedelta("0 days 00:00:01")  # threshold is the stnadard deviation of LapTime for each stint

    high_variance_stints = long_runs_with_std[long_runs_with_std['LapTime_std'] > threshold]['Stint']

    long_runs_filtered = long_runs[~long_runs["Stint"].isin(high_variance_stints)].copy()

    long_runs_filtered.reset_index(drop=True, inplace=True)

    team_long_runs_filtered[team] = long_runs_filtered
```

```python
team_stint_data = dict()

for team in all_laps_by_team:
    laps = all_laps_by_team[team]

    laps["GlobalStint"] = laps["Stint"]

    day1_laps = laps[laps["Day"] == "Day 1"]
    max_stint_day1 = day1_laps["Stint"].max()
    laps.loc[laps["Day"] == "Day 2", "GlobalStint"] += max_stint_day1

    day2_laps = laps[laps["Day"] == "Day 2"]
    max_stint_day2 = day2_laps["GlobalStint"].max()
    laps.loc[laps["Day"] == "Day 3", "GlobalStint"] += max_stint_day2


    laps = laps[laps["LapTime"] <= pd.Timedelta(minutes=2)]

    filtered_df = laps[laps["PitInTime"].isna() & laps["PitOutTime"].isna()]
    stint_counts = filtered_df["GlobalStint"].value_counts()
    print(stint_counts)
    valid_stints = stint_counts.index[stint_counts > 5]

    filtered_df = filtered_df[filtered_df["GlobalStint"].isin(valid_stints)]
    
    result = filtered_df.groupby('GlobalStint').agg(
        {'LapTime': ['mean', 'std'], 'GlobalStint': 'size'})
    team_stint_data[team] = result

    
```

```python
for team in team_stint_data:

    df = team_stint_data[team]

    df['LapTimestd_seconds'] = df['LapTimestd'].dt.total_seconds()

    weighted_std_mean = np.average(
        df['LapTimestd_seconds'], weights=df['GlobalStintsize'])

    weighted_std_mean_timedelta = pd.to_timedelta(weighted_std_mean, unit='s')

    print(f"{team} {weighted_std_mean_timedelta}")
```
